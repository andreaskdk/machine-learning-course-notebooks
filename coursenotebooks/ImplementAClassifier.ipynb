## Random Classifier

Implement a classifier that returns a random label when used for prediction

```
import random
class RandomClassifier():
  
  def fit(self, X, y):
    self.X = X
    self.y = y
    self.labels = list(set(y))
    
  
  def predict(self, X):
    return [ self.labels[random.randint(0,len(self.labels) - 1)] for i in range(len(X)) ]
```

We can evaluate this classifier on our dataset. We should expect about 33% correct labels.

```
classifier = RandomClassifier()

classifier.fit(X, y)

y_predicted = classifier.predict(X)

predicted_correct = (y == y_predicted)

sum(predicted_correct)/len(predicted_correct)
```
